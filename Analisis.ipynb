```python
## import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the dataset
data = pd.read_csv('financial_market_videogames.csv')
df = data.copy()
df.head()
```

```python
df['Company'].unique()
```

```python
df.sort_values('Date', ascending = False, inplace = True)
```

Mercado Japones

```python
jp = df[df['Currency'] == 'JPY'].copy().sort_values('Date',ascending=True)
fechas = jp['Date'] > '2015-01-01'
plt.figure(figsize = (16,10))
plt.plot(jp[fechas]['Date'],jp[fechas]['Close'],color= 'b',label='SONY Close')
plt.scatter('2021-12-01',jp[fechas]['Close'].agg(np.max), color = 'r',label = 'Max 25', s = 100)
plt.xticks(ticks = ['2015-02-01','2016-01-01','2017-01-01','2018-01-01','2019-01-01',
                    '2020-01-01','2021-01-01','2022-01-01','2023-01-01','2024-10-01'], rotation= 90)
plt.xlabel('Date')
plt.grid(color='g', linestyle='--', linewidth=0.1)
plt.ylabel('Market Close (JPY)')
plt.legend()
plt.title('Market Close Sony')
plt.show()
```

Mercado USD

```python
usd = df.copy()[np.logical_and( df['Currency'] == 'USD' , df['Date'] > '2016-12-01' )].sort_values('Date',ascending=True)
usd_grouped = usd[['Date','Close','Company']].groupby('Company')
plt.figure(figsize = (12,10))
for company, data in usd_grouped:
    plt.step(data['Date'],data['Close'],label = company)
plt.legend()
plt.xticks(ticks = ['2017-01-01','2018-01-01','2019-01-01',
                    '2020-01-01','2021-01-01','2022-01-01','2023-01-01','2024-10-01'], rotation= 90)
plt.ylabel('USD Close')
plt.xlabel('Date')
plt.grid(linewidth = 0.1)
plt.savefig('ojo.jpg')
plt.show()
```
